```python
nltk.download()
```

# open a text file

```python
with open('texto.txt', 'r') as text_file:
    lines = text_file.readlines()
```

```python
lines
```

# using tokenizer

```python
lines = [nltk.tokenize.word_tokenize(line) for line in lines]
```

```python
lines
```

# split words using spaces

```python
words = []
for line in lines:
    words.extend(line)
```

```python
words = [w.lower() for w in words]
print(words)
```

```python
stopwords = nltk.corpus.stopwords.words('portuguese')
```

```python
stopwords
```

```python
words = [w for w in words if w not in stopwords]
```

```python
words
```

# count the words grouping the equals

```python
vocab_freq = {}
for word in words:
    if word in vocab_freq:
        vocab_freq[word] += 1
    else:
        vocab_freq[word] = 1
```

```python
vocab_freq
```

# plot some statistics

distinct words
10% most frequent
10% less frequent
etc.

```python
import operator
import math

ordered_vocab_freq = sorted(vocab_freq.items(), key=operator.itemgetter(1))

ten_perc = math.floor(len(ordered_vocab_freq) / 10)

print(sorted(vocab_freq.keys()))
```

```python
for k, v in ordered_vocab_freq[:ten_perc]:
    print("{} => {}".format(k, v))
```

```python
reversed_freq = list(reversed(ordered_vocab_freq))
for k, v in reversed_freq[:ten_perc]:
    print("{} => {}".format(k, v))
```
